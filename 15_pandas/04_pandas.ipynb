## concatination of two dataframe

```python
import pandas as pd
```

```python


# Create a DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie'],  # Example names
    'sub': ['Math', 'Science', 'English'],  # Example subjects
    'mark': [85, 90, 88],
    'index':[1,2,3]
}

# Specify custom index
df = pd.DataFrame(data, index=[101, 102, 103])

print(df)
```

```python


# Create a second DataFrame
data2 = {
    'name': ['David', 'Eve', 'Frank'],  # Example names
    'sub': ['History', 'Geography', 'Physics'],  # Example subjects
    'mark': [78, 92, 89],
    'index':[1,2,3]# Example marks
}

# Specify custom index
df2 = pd.DataFrame(data2, index=[201, 202, 203])

print(df2)
```

```python
we can concat two datframe in two way 
1.horizontal
2.vartical 
```

```python
pd.concat([df,df2]) 
```

```python
pd.concat([df,df2],keys=["x","y"])
```

```python
pd.concat([df,df2],ignore_index=True) 
```

```python
pd.concat([df,df2],axis=1) 
```
