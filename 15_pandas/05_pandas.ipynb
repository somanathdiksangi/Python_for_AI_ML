## group by

```python
import pandas as pd

# Sample dataset
data = {
    'name': ['Alice', 'Bob', 'Alice', 'David', 'Bob', 'Eve', 'David', 'Eve'],
    'sub': ['Math', 'Math', 'Science', 'Math', 'Science', 'Science', 'History', 'History'],
    'mark': [85, 78, 92, 88, 81, 90, 76, 95]
}

# Create a DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
df.groupby("sub")
```

```python
df.groupby("sub").groups
```

```python
df.groupby(["sub","name"]).groups
```

```python
gr=df.groupby("sub")
```

```python
for i in gr:
    print(i)
```

```python
gr2=df.groupby("mark")
```

```python
for i in gr2:
    print(i)
```

```python
df.groupby("sub").get_group("History")
```

```python
df.groupby("sub").get_group("Science")
```

```python
df
```

```python
grp=df.groupby("sub")
```
