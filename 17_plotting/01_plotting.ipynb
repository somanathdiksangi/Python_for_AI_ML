```python
import matplotlib.pyplot as plt


x = np.random.randn(100)
y = np.random.randn(100)

# Create the figure and axes
fig, ax = plt.subplots()

# Create a scatter plot
ax.scatter(x, y)

# Set axis labels
ax.set_xlabel('x-axis')  # Corrected from set_xlable
ax.set_ylabel('y-axis')  # Corrected from set_ylable

# Set the title
ax.set_title("Scatter")  # Corrected from ax.tittle

# Show the plot
plt.show()
```
